```python
# Removing outliers for all required columns
cols = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_Q', 'Embarked_S']

X = df.drop(columns=['Survived'])
y = df['Survived']

print(X)
print(y)
```

```python
X_no_outliers = X.copy()
y_no_outliers = y.copy()

# Initialize a mask that is True for all rows
combined_mask = pd.Series(True, index=X.index)

for col in cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    mask = (X[col] >= lower_bound) & (X[col] <= upper_bound)

    # Combine mask with the overall mask using logical AND
    combined_mask = combined_mask & mask

    # Print the mask results for this column
    for idx, name in enumerate(mask.value_counts().index.tolist()):
        print(f"For {col}, {name} -> {mask.value_counts()[idx]}")

# Apply the combined mask to filter out the outliers
X_no_outliers = X_no_outliers[combined_mask]
y_no_outliers = y_no_outliers[combined_mask]
```
