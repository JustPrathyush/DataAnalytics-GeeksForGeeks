Data Imputation: Replacing all the null/missing values entries with some number.



*   Replace the missing values with the mean, median or mode
*   Apply machine learning algorithm [e.g. KNN] to predict the possible value in the replacement of missing values of a given feature.



```python
import seaborn as sns
sns.distplot(data['Age'])
```

Data Imputation of 'age' with the mean or median [numeric/int data]

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
## data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
data.isnull().sum()
```

```python
sns.distplot(data['Age'])
```

Data Imputation of 'Embarked' feature in a given dataset

```python
data['Embarked'].mode()[0]
```

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```
